# Aufgabe 5:
Aufgabenstellung:
Entwickeln Sie ein Python-Programm, das die Volumenberechnung eines Quaders durchfhrt und zustzlich erkennt, ob es sich bei dem Krper um einen speziellen Quader - nmlich einen Wrfel - handelt.
Anforderungen:

* Das Programm soll den Benutzer nach drei Seitenlngen fragen.
* Die Eingaben sollen als Gleitkommazahlen (float) verarbeitet werden.
* Das Programm berechnet das Volumen des Krpers.
* Das Programm soll berprfen, ob alle drei Seiten gleich lang sind.
* Je nach Ergebnis soll das Programm ausgeben:
  
        "Das Volumen des Wrfels ist gleich [Volumen]" (wenn alle Seiten gleich sind)
        "Das Volumen des Quaders ist gleich [Volumen]" (wenn mindestens zwei Seiten unterschiedlich lang sind)


````python
seite1 = float(input("Seitenlnge 1:"))
seite2 = float(input("Seitenlnge 2:"))
seite3 = float(input("Seitenlnge 3:"))
volumen = seite1 * seite2 * seite3
if seite1 == seite2 and seite2 == seite3:
    print("Das Volumen des Wrfels ist gleich", volumen)
else:
    print("Das Volumen des Quaders Wrfels ist gleich", volumen)
````

**Erweiterung (optional):**

Erweitern Sie das Programm um eine Prfung, die sicherstellt, dass nur positive Werte fr die Seitenlngen eingegeben werden knnen.

````python
seite1 = float(input("Seitenlnge 1:"))
seite2 = float(input("Seitenlnge 2:"))
seite3 = float(input("Seitenlnge 3:"))
if seite1 <= 0 or seite2 <= 0 or seite3 <= 0:
    print("Falsche Eingabe. Die Seitenlngen sollen > 0 sein.")
else:
    volumen = seite1 * seite2 * seite3
    if seite1 == seite2 and seite2 == seite3:
        print("Das Volumen des Wrfels ist gleich", volumen)
    else:
        print("Das Volumen des Quaders Wrfels ist gleich", volumen)
````
