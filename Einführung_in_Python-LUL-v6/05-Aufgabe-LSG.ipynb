# Aufgabe:
Entwerfen Sie ein Programm, das die Rechenoperationen plus, minus, mal, geteilt und hoch durchführt. Definieren Sie dazu zwei Variablen, die Sie für die Berechnungen nutzen und geben Sie das Ergebnis am Bildschirm aus.
### Problemanalyse:
#### Welche Ausgabedaten will man erhalten?

Das Ergebnis von Berechnungen: Summe, Differenz, Produkt, Quotient, Potenz

#### Welche Daten werden zur Bearbeitung benötigt?

Zwei Zahlen

#### Welche Eigenschaften haben die Ausgabedaten? (Variablenliste)

|Bedeutung| Inhalt| Variablenname|
|-|-|-|
|Erste Zahl| Ganzzahl| zahl1|
|Zweite Zahl| Ganzzahl| zahl2|
|Summe| Ganzzahl| addition|
|Differenz| Ganzzahl| subtraktion|
|Produkt |Ganzzahl |multiplikation|
|Quotient| Kommazahl| division|
|Potenz |Ganzzahl |potenz|

#### So soll die Bildschirmausgabe des Programms aussehen (Beispiel):
```
Zahl1: 13, Zahl2: 7
---------------------
Addition: 20
Subtraktion: 6
Multiplikation: 91
Division: 1.8571428571
Potenz: 62748517
```
### Struktogramm:
Zeichnen Sie ein Struktogramm zu diesem Programm.

### Programmcode (Python-Code)

````python
#Hier den Programmcode eingeben und mit STRG+ENTER ausführen
zahl1 = 13
zahl2 = 7

# Durchführen aller Berechnungen:
addition = zahl1 + zahl2
subtraktion = zahl1 - zahl2
multiplikation = zahl1 * zahl2
division = zahl1 / zahl2
potenz = zahl1 ** zahl2

# Ausgabe
# Überschrift
print("Zahl1:", zahl1, "Zahl2:", zahl2) 
print("-------------------")

# Ergebnisse
print("Addition:", addition)
print("Subtraktion:", subtraktion)
print("Multiplikation:", multiplikation)
print("Division:", division)
print("Potenz:", potenz)
````
