# Funktionen mit Parameter
Einem Unterprogramm können Werte beim Aufruf übergeben werden.   
Bei der Definition des Unterprogramms gibt man die Werte zwischen den runden Klammern nach dem Namen an. 

```python
def add(a,b):
    x=a+b
    print("Ergebnis:", x)
```

```python
add(3,4)
```
