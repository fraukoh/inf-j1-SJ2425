# Aufgabe: Mietzuschuss

Schreiben Sie ein Programm, das den Mietzuschuss in Abhängigkeit vom Mietpreis berechnet. Bei einer Miete von weniger als 500,00 Euro beträgt der Zuschuss 2%. Von 500,00 Euro bis unter 1000,00 Euro beträgt der Zuschuss 5% und ab 1000,00 Euro erhält man einen Zuschuss von 7%.

```python
miete = float(input("Bitte Miete eingeben: "))
if miete < 500:
    zuschuss = miete * 0.02    
else:
    if miete < 1000:
        zuschuss = miete * 0.05        
    else:
        zuschuss = miete * 0.07       
print("Der Mietzuschuss beträgt:", zuschuss)
```

# Aufgabe: Anmeldung berufliches Gymnasium

In diesem Projekt soll überprüft werden, ob eine Anmeldung mit einem mittleren Bildungsabschluss an einem beruflichen Gymnasium in Baden-Württemberg möglich ist oder nicht. Die Bedingungen dafür lauten wie folgt: Abschlusszeugnis (Mittlerer Bildungsabschluss) der Realschule, der zweijährigen Berufsfachschule, der Berufsaufbauschule oder der Werkrealschule mit einem Notendurchschnitt von mindestens 3,0 aus Deutsch, Mathematik und Englisch (kein Fach schlechter als Note ausreichend).

```python
note_deutsch = int(input("Deutschnote: "))
note_englisch = int(input("Englischnote: "))
note_mathematik = int(input("Mathematiknote: "))
ist_zugelassen = False

if note_deutsch <= 4:
    if note_englisch <= 4:        
        if note_mathematik <= 4:
            if (note_deutsch + note_englisch + note_mathematik) / 3 <= 3:
                ist_zugelassen = True
if ist_zugelassen == True:
    print("Die Aufnahme ist möglich.")
else:
    print("Aufnahme ist nicht möglich.")
```
